```python
"""
Kafka 컨슈머의 수동 오프셋 커밋을 실습하는 코드입니다.

TODO:
1. kafka-python 라이브러리를 사용하여 KafkaConsumer를 생성합니다.
2. 컨슈머의 자동 오프셋 커밋을 비활성화(`enable_auto_commit=False`)합니다.
3. 5개씩 끊어서 총 3번 메시지를 소비하는데, 3번째 실행에서는 커밋하지 않습니다.
4. 다시 메시지를 소비하여 마지막 커밋된 위치부터 읽는 것을 확인합니다.
"""

from kafka import KafkaConsumer
```

```python
# TODO 1: consume_messages 함수를 구현하세요.
def consume_messages(iteration, commit=True):
    """ 특정 배치만큼 메시지를 소비하고 선택적으로 커밋하는 함수 """

    # TODO: KafkaConsumer 인스턴스를 생성하세요.
    consumer = KafkaConsumer(
        "test-topic",  # 구독할 토픽 설정
        bootstrap_servers='localhost:9092',  # Kafka 브로커 주소 설정
        auto_offset_reset='earliest',  # 메시지를 처음부터 읽을지 설정
        enable_auto_commit=True,  # 자동 오프셋 커밋 설정
        group_id='my-group'  # 컨슈머 그룹 ID 설정
    )

    print(f"\n=== Batch {iteration} ===")
    messages = []

    for _ in range(5):
        message = next(consumer)  # 컨슈머에서 메시지를 하나씩 가져옴
        messages.append(message.value.decode('utf-8'))

    print("Received Messages:", messages)

    if commit:
        consumer.commit()  # 수동 오프셋 커밋하는 함수
        print("오프셋 커밋 완료")

    consumer.close()  # 컨슈머 종료하는 함수
```

```python
# TODO 2: 5개씩 3번 메시지를 소비하되, 마지막 실행에서는 커밋하지 않습니다.
for i in range(2):
    consume_messages(i + 1, commit=True)  # 메시지 소비 후 오프셋 커밋 여부 설정

consume_messages(3, commit=False)  # 마지막 실행에서는 커밋하지 않음
```

```python
# TODO 3: 다시 메시지를 소비하여 마지막 커밋된 위치부터 읽기
consume_messages(4, commit=True)  # 마지막 커밋된 위치부터 메시지를 소비하고 오프셋 커밋 여부 설정
```
